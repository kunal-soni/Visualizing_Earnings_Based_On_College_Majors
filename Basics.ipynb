# Observations from the scatter plots

- Sample size vs Sample size - The highest bar is for lowest sample size (0-400) indicating that there are more frequency of sample size being low.
- Sample size vs Median - There is more median concentrated between 30000 and 40000 and the sample size for that area is on lower side (0-500)
- Sample size vs Unemployment rate - There are darker scatter plots between 0.025 to 0.10 with most of them for low sample size (0-400)
- Median vs Median - The tallest bar is for 35000 - 40000 median income indicating more frequency for this salary range
- Median vs Unemployment rate - There is no particular area with darker scatter points indicating that unemployment rate can occur for anywhere between lower, medium or high salary.
- Unemploment rate vs Unemployment rate - The highest bar is for 0.06 - 0.08 indicating mostly unemployment is between 6-8%, but the bars for 4-12% are higher so there is more confidence of unemployment being in this range

```python
# Use bar plots to compare percentages of women (ShareWomen) from first ten 
# and last ten rows
recent_grads[:10].plot.bar(x="Major",y="ShareWomen")
```

```python
recent_grads[-10:].plot.bar(x="Major",y="ShareWomen")
```

# Percentages of women in first and last ten rows
The above bar charts show that in top ten ranked college majors, women have lower percentage. Most are below 20%. Three of the ten have between 30-55%
(Chemical Engineering, Actuarial Science and Astronomy and Astrophysics).

For the last ten rows (lower ranked college majors), there are two with 100% women (Communication Disorders Sciences and Services, and Early Childhood Education), and all are above 60%.

```python
# Use bar plots to compare unemployment rate from first ten and last ten rows
recent_grads[:10].plot.bar(x="Major",y="Unemployment_rate")
```

```python
recent_grads[-10:].plot.bar(x="Major",y="Unemployment_rate")
```

# Unemployment rate for top ten and bottom ten ranked college majors

The above bars show there is not significatn difference in unemployment rate between top ranked and bottom ranked college majors. 

The highest ranked major has only 2% unemployment rate (Petroleum Engineering), but the 2nd (Mining and Mineral Engineering) and 6th ranked (Nuclear Engineering) majors have 12% and 18% which is higher than average. For the bottom ranked majors, the unemployment rate varies between 4% (Early childhood education) and 15% (Clinical Psychology).
